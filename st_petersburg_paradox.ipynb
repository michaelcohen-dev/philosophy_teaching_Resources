# The St. Petersburg Paradox

The following piece of code simulates the St. Petersburg paradox. You can read about the paradox in chapter 14 of the book *Odds and Ends* by Jonathan Weisberg, [available here](https://https://jonathanweisberg.org/vip/infinity-beyond.html#infinity-beyond).

image.png


```python
import random


RUNS = 100000

results = []

max_rounds = 0

for i in range(RUNS):

  rounds = 1

  # Here 1 is tails, so if we get 0 at start,
  # this means we got heads right away and we
  # are done with the game
  while random.randint(0,1):

    rounds += 1

  winning_value = 2 ** rounds

  if rounds > max_rounds: max_rounds = rounds

  results.append(winning_value)


print("average (mean) winnings after ", RUNS, " runs:")
print(sum(results)/float(RUNS))
print("Among the runs, the longest game had ", max_rounds ," rounds")





```
