# Condorcet's jury theorem simple simulation


Here is a simple formulation of the Condorcet's jury theorem:

* A group of voters (jurors) decides on accepting a claim by a majority vote.
* Each voter makes an independent decision about the claim.
* Let p be the probability that each member gets it right.

Then:
* If p > 0.5, then the larger the group, the more likely that they will get the right decision
* If p > 0.5, then the larger the group, the more likely that they will  get the wrong decision.

In even simple terms, if each voter is more likely to vote correctly than not, the the more voters you add, the more likely it is for the vote to get the correct decision.

Below is a simple simulation of this sensrio (not a proof, BTW).

```python
import random

nums_of_jurors = [10, 100, 1000, 10000, 100000]
prob_of_truth = 0.501
RUNS = 100

def generate_binary_list(n, p):
    binary_list = [1 if random.random() < p else 0 for _ in range(n)]
    return binary_list

for number in nums_of_jurors:
  correct_vote = 0
  for i in range(RUNS):

    current_votes = generate_binary_list(number, prob_of_truth)

    if (sum(current_votes) > ( number / 2)):
      correct_vote +=1

  print("among the ", RUNS, " runs of the simulation")
  print("the ", number, " jurors got the vote right ", correct_vote, " times")
  print("===============================")



```
